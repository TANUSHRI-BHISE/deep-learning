```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts([faqs])
```

```python
tokenizer.word_index
```

```python
unique_words=len(tokenizer.word_index)+1#as its indexing starts from 1 and 0 is for padding
```

```python
input_sequences=[]
for sentences in faqs.split('\n'):#from new line split the text #sentence=we have each sentence
  tokenized_sentences=(tokenizer.texts_to_sequences([sentences])[0])#convert text to number [sentences] list used as more than one list can be passed [sentences])[0]used to remove double list
  for i in range(1,len(tokenized_sentences)):#eg 1st sentence is of 3 words hence loop from 1 to 3
    input_sequences.append(tokenized_sentences[:i+1])#from start to start +1 is picked
    #eg 1st semtence is i love india
     # |     input   |     output   |
    #  1)i           1) love
    #  2)i love      2)  india
# we are extracting
```

```python
input_sequences
```

```python
max_length=max([len(x) for x in input_sequences])#which sentence has max words as we want it for 0 padding
```

```python
max_length
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
padded_input_sequence=pad_sequences(input_sequences,maxlen=max_length,padding='pre')
```

```python
padded_input_sequence
```

```python
X=padded_input_sequence[:,:-1]
```

```python
y=padded_input_sequence[:,-1]
```

```python
X.shape
```

```python
y.shape
```

```python
len(tokenizer.word_index)# we have total 282 unique words
#56 is the largest length of a sentence
```

```python
# y is a single number and its classification task not regression hence we need to do one hot encoding of it
```

```python
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=283)#case index starts from 1 in tokenize.word_index but actullay its from 0
# num_classes= no of unique words
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense,Input
```

```python
model = Sequential()
model.add(Input(shape=(56,)))# 1 sentence =56 words hence for 283 uniquew words we have 283*100 outputs from embedding layer
model.add(Embedding(283, 100, input_length=56))#283 unique words ,100 output for single word ,max_input length =max length of single word
model.add(LSTM(150))#nodes in lstm layer
# model.add(LSTM(150))
model.add(Dense(283, activation='softmax'))# its multi class classification hence we want probablity of each word who ever have higest probablity wins
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
 #1st layer# 1 sentence =56 words hence for 283 uniquew words we have 283*100 outputs from embedding layer
 #2st layer#in lstm layer 56 word each sentence of 100 dimension
 #3 layer # 283 nodes do multiclass classifiction for each word and higest probablity word is retrived
```

```python
model.fit(X,y,epochs=100)
```

```python
import time
text = "what is the fee"

for i in range(10):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=56, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      time.sleep(2)
```
