```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input,decode_predictions
import numpy as np
```

```python
model=ResNet50(weights='imagenet')
```

```python
img_path='/content/sample_data/tomato.jpg'
img=image.load_img(img_path,target_size=(224,224))
x=image.img_to_array(img)#convert to aaray
x=np.expand_dims(x,axis=0)#batch me convert
x=preprocess_input(x)#preprocess kar ke x me dal diya
```

```python
preds=model.predict(x)
print('predicted',decode_predictions(preds,top=3)[0])#make predictions and show top 3 prediction
```
