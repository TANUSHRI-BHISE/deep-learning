```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense,LSTM,GRU,Bidirectional
```

```python
# Load the IMDb dataset
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=10000)

# Pad sequences to have the same length
x_train = pad_sequences(x_train, maxlen=100,padding='post',truncating='post')
x_test = pad_sequences(x_test, maxlen=100,padding='post',truncating='post')
```

```python
model=Sequential([
    Embedding(input_dim=10000, output_dim=32, input_shape=(100,)),#unique word ,output against each words,words in each sentences
    SimpleRNN(5),#rey=turn sequences is true beacuse we want to send output of each to the next hidden layer so dont do the same for last layer
    Dense(1,activation='sigmoid')

])
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model=Sequential([
    Embedding(input_dim=10000, output_dim=32, input_shape=(100,)),#unique word ,output against each words,words in each sentences
    Bidirectional(SimpleRNN(5)),#we used simple rnn in bidirectional wrapper
    Dense(1,activation='sigmoid')

])
model.summary()
```

```python
#in bidirectional layer weights and biases doubles ie 2 times of simple rnn
```

```python
model=Sequential([
    Embedding(input_dim=10000, output_dim=32, input_shape=(100,)),#unique word ,output against each words,words in each sentences
    Bidirectional(LSTM(5)),#we used simple rnn in bidirectional wrapper
    Dense(1,activation='sigmoid')

])


model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.summary()
```

```python
model=Sequential([
    Embedding(input_dim=10000, output_dim=32, input_shape=(100,)),#unique word ,output against each words,words in each sentences
    Bidirectional(GRU(5)),#we used simple rnn in bidirectional wrapper
    Dense(1,activation='sigmoid')

])


model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.summary()
```

```python
history=model.fit(x_train,y_train,epochs=5,batch_size=32, validation_split=0.2)
```
