```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

```python
docs = ['go india',
		'india india',
		'hip hip hurray',
		'jeetega bhai jeetega india jeetega',
		'bharat mata ki jai',
		'kohli kohli',
		'sachin sachin',
		'dhoni dhoni',
		'modi ji ki jai',
		'inquilab zindabad']
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer(oov_token='<nothing>')
```

```python
tokenizer.fit_on_texts(docs)
```

```python
tokenizer.document_count
```

```python
sequences = tokenizer.texts_to_sequences(docs)
sequences
```

```python
from tensorflow.keras.utils import pad_sequences
```

```python
from tensorflow.keras.datasets import imdb
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,SimpleRNN,Embedding,Flatten,Input
```

```python
(X_train,y_train),(X_test,y_test)=imdb.load_data()
```

```python
#its a already preprocessed data set availabel in keras
```

```python
X_train[0]
```

```python
len(X_train[0])
```

```python
len(X_train[22])
```

```python
#all have diffrent lengths#   ie 22nd review have 256 woerds in it
```

```python
X_train=pad_sequences(X_train,padding='post',maxlen=50)#use only 50 words
X_test=pad_sequences(X_test,padding='post',maxlen=50)
```

```python
X_train.shape
```

```python
model=Sequential()
model.add(SimpleRNN(32,input_shape=(50,1),return_sequences=False))#32 nodes shape 50(time straps=no of words) ,1(at a time 1 word input) return seq=false means normal flow to of rnn if its true means output flows out (towrd programmer)at each level
model.add(Dense(1,activation='sigmoid'))#sentimental analysis is binary classification
model.summary()
```

```python
model.compile(loss="binary_crossentropy",optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```

```python
#if more number of words then padding may create sparse matric but here we want dense metric so we use embedding
```

```python
#we can use enbedding layer which will create a dense metric
```

```python
model=Sequential()
model.add(Embedding(17,output_dim=2,input_length=5))#(17=total unique words,2=no of output words we want(how each many types each word to be represented),5=[9,1,0,0,0] #how each sentence is represeneted as 5 bits that is how many words in each input)
#17 input nodes giving 2 output, for every word we get 2 words for 5 words
model.summary()
```

```python
#for imdb data set
model=Sequential()
model.add(Input(shape=(50,)))
model.add(Embedding(input_dim=90000,output_dim=2,input_length=50))
model.add(SimpleRNN(32,return_sequences=False))
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```
