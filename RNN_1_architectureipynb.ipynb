```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
```

```python
from keras.layers import Dense,SimpleRNN
```

```python
model=Sequential()
```

```python
model.add(SimpleRNN(3,input_shape=(4,5)))#4 is time strap as 4 unique words ,5 is encoded form of word[10000] represents hi
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
print(model.get_weights()[0].shape)
```

```python
model.get_weights()#3*5 of 1st input layer
```

```python
print(model.get_weights()[1].shape)
```
