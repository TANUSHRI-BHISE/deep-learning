```python

from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img=image.load_img('di.jpg',target_size=(200,200),keep_aspect_ratio=True)
```

```python
datagen=ImageDataGenerator(
    rotation_range=30,#angle
    shear_range=0.2,#distortion
    zoom_range=0.2,
    horizontal_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2,
    fill_mode='reflect'#we can also use constant to fill black
    
)
```

```python
img=image.img_to_array(img)
```

```python
type(img)
```

```python
img.shape
```

```python
input_batch=img.reshape(1,200,200,3)#1 specify batch of images given 1 means batch inclde only 1 image
```

```python
i=0                        #input me 1 image,output me aur ek batch bane ga,aug nam ki directory me save kiya hai
for output in datagen.flow(input_batch,batch_size=1,save_to_dir='aug'):#flow used when only 1 image flow_from_directory used for all images in directory 
    i=i+1
    if i==10:
        break

#aaise 10 images bane ge
```
