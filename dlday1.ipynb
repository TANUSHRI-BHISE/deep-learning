```python
df.drop(columns=["RowNumber","CustomerId","Surname"],inplace=True)
```

```python
df.head()
```

```python
df=pd.get_dummies(df,columns=['Geography','Gender'],drop_first=True)
```

```python
df=df.astype(int)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(columns=['Exited'])
y=df['Exited']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(3,activation='sigmoid',input_dim=11))#input layer me 3 node hai and 11 inputs ie columns 11*3 weight +3 biases
model.add(Dense(1,activation='sigmoid'))#output layer me 1 layer 3*1 weights + 1 bias
```

```python
model.summary()
```

```python
#complie the model
model.compile(loss='binary_crossentropy',optimizer='Adam')
```

```python
model.fit(X_train_scaled,y_train,epochs=10)
```

```python
model.layers[0].get_weights()#we get 33 weights and 3 bias
```

```python
model.layers[1].get_weights()
```

```python
y_log=model.predict(X_test_scaled)
#if >0.5 then 1 else 0
```

```python
np.where(y_log>0.5,1,0)
```

```python
y_pred=np.where(y_log>0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
model=Sequential()
model.add(Dense(11,activation='relu',input_dim=11))
model.add(Dense(11,activation='relu'))
model.add(Dense(1,activation='sigmoid'))#output layer me 1 layer 3*1 weights + 1 bias
```
