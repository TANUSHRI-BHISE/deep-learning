```python
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.datasets import imdb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,SimpleRNN,Dense,LSTM,GRU
```

```python
model=Sequential([
    Embedding(input_dim=10000, output_dim=32, input_shape=(100,)),#unique word ,output against each words,words in each sentences
    SimpleRNN(5,return_sequences=True),#rey=turn sequences is true beacuse we want to send output of each to the next hidden layer so dont do the same for last layer
    SimpleRNN(5),
    Dense(1,activation='sigmoid')

])
model.summary()
```

```python
model=Sequential([
    Embedding(input_dim=10000, output_dim=32, input_shape=(100,)),#unique word ,output against each words,words in each sentences
    LSTM(5,return_sequences=True),#rey=turn sequences is true beacuse we want to send output of each to the next hidden layer so dont do the same for last layer
    LSTM(5),
    Dense(1,activation='sigmoid')

])
model.summary()
```

```python
model=Sequential([
    Embedding(input_dim=10000, output_dim=32, input_shape=(100,)),#unique word ,output against each words,words in each sentences
    GRU(5,return_sequences=True),#rey=turn sequences is true beacuse we want to send output of each to the next hidden layer so dont do the same for last layer
    GRU(5),
    Dense(1,activation='sigmoid')

])
model.summary()
```
